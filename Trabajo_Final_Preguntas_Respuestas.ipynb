### Tokenizadores

Ahora, utilizará el objeto [tokenizer](https://huggingface.co/transformers/main_classes/tokenizer.html) de Hugging Face. Podemos cargar un tokenizador usando diferentes métodos. En este caso vamos utilizar el tokenizador que creó  Hugging Face. Con este tokenizador, puedemos asegurarse de que los tokens que obtenga para el conjunto de datos coincidan con los tokens utilizados en la implementación original de BERT.

```python
# Import the AutoTokenizer from the transformers library
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('mrm8488/distill-bert-base-spanish-wwm-cased-finetuned-spa-squad2-es')
```

```python
def process_samples(sample):
    context = sample['context']
    question = sample['question']

    # Verificar si la clave 'answers.answer_start' existe y no está vacía
    if 'answers.answer_start' in sample and sample['answers.answer_start']:
        answer_start_byte = sample['answers.answer_start'][0]
        answer_end_byte = answer_start_byte + len(sample['answers.text'][0])
    else:
        # Si la clave no existe o la lista está vacía, asignar valores predeterminados o especiales
        answer_start_byte = -1
        answer_end_byte = -1

    # Tokenizar el contexto y la pregunta
    tokenized_data = tokenizer(context, question, truncation="only_first", padding="max_length", max_length=512)

    # Verificar si la clave 'answers.text' existe y no está vacía
    if 'answers.text' in sample and sample['answers.text']:
        # Encontrar el índice del token que corresponde al inicio y fin de la respuesta
        start_char = context.find(sample['answers.text'][0])
        end_char = start_char + len(sample['answers.text'][0])

        # Encontrar los índices de los tokens correspondientes al inicio y fin de la respuesta
        start_token = tokenized_data.char_to_token(start_char)
        end_token = tokenized_data.char_to_token(end_char - 1)

        # Si el inicio o fin de la respuesta no está en el tokenizado, asignar el token especial [PAD]
        if start_token is None:
            start_token = tokenizer.pad_token_id
        if end_token is None:
            end_token = tokenizer.pad_token_id
    else:
        # Si no hay respuesta, asignar el token especial [PAD]
        start_token = tokenizer.pad_token_id
        end_token = tokenizer.pad_token_id

    return {
        'input_ids': tokenized_data['input_ids'],
        'attention_mask': tokenized_data['attention_mask'],
        'start_positions': start_token,
        'end_positions': end_token
    }
```

```python
processed_train_data = flattened_train_data.map(process_samples)
processed_test_data = flattened_test_data.map(process_samples)
```

# Transformers

```python
# Cargamos el modelo pre-entrenado específico para responder preguntas.
from transformers import AutoModelForQuestionAnswering
model = AutoModelForQuestionAnswering.from_pretrained("mrm8488/distill-bert-base-spanish-wwm-cased-finetuned-spa-squad2-es")
```

```python
# Formateamos los conjuntos de datos de entrenamiento y prueba
columns_to_return = ['input_ids','attention_mask', 'start_positions', 'end_positions']
processed_train_data.set_format(type='pt', columns=columns_to_return)
processed_test_data.set_format(type='pt', columns=columns_to_return)
```

```python
# Funcion para calcular las métricas de puntuación F1 y evaluar el rendimiento del modelo
from sklearn.metrics import f1_score

def compute_f1_metrics(pred):
    start_labels = pred.label_ids[0]
    start_preds = pred.predictions[0].argmax(-1)
    end_labels = pred.label_ids[1]
    end_preds = pred.predictions[1].argmax(-1)

    f1_start = f1_score(start_labels, start_preds, average='macro')
    f1_end = f1_score(end_labels, end_preds, average='macro')

    return {
        'f1_start': f1_start,
        'f1_end': f1_end,
    }
```

```python
from transformers import Trainer, TrainingArguments
# Configuración de los argumentos de entrenamient0
training_args = TrainingArguments(
    output_dir='model_results5',          # output directory
    overwrite_output_dir=True,
    num_train_epochs=3,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=8,   # batch size for evaluation
    warmup_steps=20,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir=None,            # directory for storing logs
    logging_steps=50
)

trainer = Trainer(
    model=model, # the instantiated 🤗 Transformers model to be trained
    args=training_args, # training arguments, defined above
    train_dataset=processed_train_data, # training dataset
    eval_dataset=processed_test_data, # evaluation dataset
    compute_metrics=compute_f1_metrics
)

trainer.train()
```

```python
# Evaluacion del modelo utilizando el conjunto de datos de prueba
trainer.evaluate(processed_test_data)
```

```python
import torch
text = r"""
Puno es una ciudad ubicada en el sur de Perú, en la meseta de Collao, a una altitud de aproximadamente 3 827 metros sobre el nivel del mar. Es la capital de la región homónima de Puno y se encuentra a orillas del famoso lago Titicaca, el lago navegable más alto del mundo y uno de los cuerpos de agua más grandes de Sudamérica.
"""

questions = ["¿donde esta ubicado puno?","¿Donde se encuentra? "]

for question in questions:
    inputs = tokenizer.encode_plus(question, text, return_tensors="pt")
    input_ids = inputs["input_ids"].tolist()[0]
    inputs.to("cuda")

    text_tokens = tokenizer.convert_ids_to_tokens(input_ids)
    answer_model = model(**inputs)

    answer_start = torch.argmax(
        answer_model['start_logits']
    )  # Get the most likely beginning of answer with the argmax of the score
    answer_end = torch.argmax(answer_model['end_logits']) + 1  # Get the most likely end of answer with the argmax of the score

    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))

    print(f"Question: {question}")
    print(f"Answer: {answer}\n")
```

```python
import fitz  # Importar módulo PyMuPDF (Fitz)
from google.colab import files

# Cargar archivo PDF
pdf_file = files.upload()

# Nombre del archivo PDF cargado
pdf_filename = list(pdf_file.keys())[0]

# Abrir el archivo PDF
pdf_document = fitz.open(pdf_filename)

# Verificar si el archivo PDF tiene al menos una página
if pdf_document.page_count > 0:
    # Extraer el texto de la primera página del PDF
    page = pdf_document.load_page(0)
    pdf_text = page.get_text()

    # Guardar el texto en un archivo TXT
    txt_filename = pdf_filename.replace(".pdf", ".txt")
    with open(txt_filename, "w") as txt_file:
        txt_file.write(pdf_text)
```
